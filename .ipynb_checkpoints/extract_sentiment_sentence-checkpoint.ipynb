```python
import codecs
from functools import reduce
import jieba
import time
import re
```

```python
with codecs.open('sentiment_word/positive_word.txt','r',encoding = 'utf-8') as fp,\
     codecs.open('sentiment_word/negative_word.txt', 'r', encoding = 'utf-8') as fn:
    pos_words = [lines.strip() for lines in fp]
    neg_words = [lines.strip() for lines in fn]
```

```python
print('pos_word : %d' % len(pos_words))
print('neg_word : %d' % len(neg_words))
```

```python
start = time.clock()
with codecs.open('/data/xuman/sentiment_analysis/allmsg/save_msgs_thr5.txt', 'r') as fr,\
     codecs.open('/data/xuman/sentiment_analysis/allmsg/msg_pos', 'w') as fw_pos,\
     codecs.open('/data/xuman/sentiment_analysis/allmsg/msg_neg', 'w') as fw_neg:
   
    lines = [x for x in fr ]
    s_pos= re.compile(r"(你真)|棒|赞|(好评)|(五星)|(满意)|(不错)|(喜欢)|(厉害)")
    s_neg = re.compile(r"(差评)|傻|(妈的)|(去死)|(滚)|(讨厌)|烦|(看不上)")
    line_pos = [ x for x in lines if s_pos.findall(str(x))]
    line_neg = [ x for x in lines if s_neg.findall(str(x))]

    _ = list(map(lambda x: fw_pos.write(str(x)),line_pos)) 
    _ = list(map(lambda x: fw_neg.write(str(x)),line_neg))
      
end = time.clock()
print('time used %8f' % (end - start))
```

```python
with codecs.open('/data/xuman/sentiment_analysis/allmsg/msg_pos_simhash.txt', 'r') as fw_pos,\
     codecs.open('/data/xuman/sentiment_analysis/allmsg/msg_pos_remove.txt','w') as fw_pos_rm, \
     codecs.open('/data/xuman/sentiment_analysis/allmsg/msg_neg', 'r') as fw_neg:
    s1 = re.compile(r"\[语音解析\]")
    lines = [s1.sub('',x) for x in fw_pos ] 
    for line in lines:
        if len(line) >= 16:
            lines.remove(line)
        else:
            line = re.compile(r"\d{1,2}月\d{1,2}[日 号]").findall(line)
            line = re.complie(r"提醒|麻烦|为什么")
            line = re.compile(r"00.*|@.*")
            line = re.compile(r"\d{1,2}[分 号 点 日]")
            line = re.compile(r"\d{1,2}\S.*")
            line = re.compile(r"\S{3,}我喜欢")
            line = re.compile(r"他.*")
            line = re.compile(r"五|帮|她|你喜欢")
             
#     s2 = re.compile(r"(\d{1,2}月\d{1,2}[日 号])|提醒|麻烦|为什么|00.*|@.*|\d{1,2}[分 号 点 日]|\d{1,2}\S.*|他.*|\S{3,}我喜欢|五|帮|她|你喜欢")
#     lines = [line for line in lines if  not s2.findall(line) and len(line) < 16]
    _ = list(map(lambda x: fw_pos_rm.write(str(x)),lines))
    print('length: %d \n' % len(lines),'\n')
    for x in lines:
        print((x))

```

```python
num_1 ,num_2_4, num_5_8, num_8_16, num_16_ = 0,0,0,0,0
with open('/data/xuman/sentiment_analysis/allmsg/msg_pos','r') as f:
    lines = f.readlines()
    le = len(lines)
    print(len(lines))
    for line in lines:
        if len(line) == 1:
            print('dd')
            num_1 += 1
        elif 2 <= len(line) <= 4:
            num_2_4 +=1
            
        elif 5 <= len(line) <= 8:  
            num_5_8 += 1
        elif 8 <= len(line) <= 16: 
            num_8_16 += 1
        elif len(line) > 16:
            num_16_ += 1
print('num_1 %d \t '% num_1,'1_ratio %3f \t' % (num_1/le*100),'\n')
print('num_2_4 %d \t'% num_2_4,'\t','1_ratio %3f \n' % (num_2_4/le*100),'\n')
print('num_5_8 %d \t'% num_5_8,'\t','num_5_8 %3f \n' % (num_5_8/le*100),'\n')
print('num_8_16 %d \t '% num_8_16,'\t','1_ratio %3f \n' % (num_8_16/le*100),'\n')
print('num_16_ %d \t '% num_16_,'\t','1_ratio %3f \n' % (num_16_/le*100),'\n')
```
